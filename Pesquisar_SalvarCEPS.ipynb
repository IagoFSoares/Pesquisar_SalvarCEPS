```python
#importando biblioteca openpyxl
from openpyxl import load_workbook
import os

#Pegamos o caminho do arquivo do excel e mais seu nome
nomeArquivoCEP = 'C:\\Users\\ADMIN\\Documents\\Python RPA e Excel como automatizar Processos e Planilhas!\\SEÇAO 4 - ROBÔ 3 - ESTRAINDO CEPS PELO SITE CORREIOS\\PesquisaEndereco_2.xlsx'
planilhaDadosEndereco = load_workbook(nomeArquivoCEP)

#selecionando a planilha/sheet de CEP
sheetSelecionada = planilhaDadosEndereco["CEP"]

#importando as bibliotecas selenium e pyautogui
from selenium import webdriver as opcoesSelenium
from selenium.webdriver.common.keys import Keys
import pyautogui as tempoEspera

#importando elemento By
from selenium.webdriver.common.by import By

#abre o navegador do Chrome e abre o site de busca cep
navegador = opcoesSelenium.Chrome()
navegador.get('https://buscacepinter.correios.com.br/app/endereco/index.php')

#tempo de espera
tempoEspera.sleep(2)

#imprime o CEP no campo(que pode ser pelo elemonto Name, Id ou até X.PATH)
navegador.find_element(By.NAME, 'endereco').send_keys('33900020')
#navegador.find_element(By.XPATH, //*[@id="endereco"]).send_keys('32223220')
#navegador.find_element(By.ID, 'endereco').send_keys('32223220')

#tempo de espera
tempoEspera.sleep(2)

#clicando no botão btn_pesquisar para pesquisar
navegador.find_element(By.NAME, 'btn_pesquisar').click()

#tempo de espera
tempoEspera.sleep(2)

#para 
for linha in range(2, len(sheetSelecionada['A']) + 1):
    
    #tempo de espera
    tempoEspera.sleep(2)
    
    #pesquisando pelo botão nova busca através do elemento ID(ou XPATH) e clicando para fazer nova busca
    navegador.find_element(By.ID, 'btn_nbusca').click()
    #navegador.find_element(By.ID, '//*[@id="btn_nbusca"]').click()
    
    #tempo de espera
    tempoEspera.sleep(2)
    
    #pegando o  cep da planilha e colocando na variável 
    cepPesquisa = sheetSelecionada['A%s' % linha].value
    
    #tempo de espera
    tempoEspera.sleep(2)
    
    #imprime o CEP no campo(que pode ser pelo elemonto Name, Id ou até X.PATH)
    navegador.find_element(By.NAME, 'endereco').send_keys(cepPesquisa)

    #tempo de espera
    tempoEspera.sleep(2)

    #clicando no botão btn_pesquisar para pesquisar
    navegador.find_element(By.NAME, 'btn_pesquisar').click()

    #tempo de espera
    tempoEspera.sleep(2)
    
    #pega os dados da rua pelo X.PATH
    rua = navegador.find_elements(By.XPATH, '//*[@id="resultado-DNEC"]/tbody/tr/td[1]')[0].text
    print(rua)

    #pega os dados do bairro pelo X.PATH
    bairro = navegador.find_elements(By.XPATH, '//*[@id="resultado-DNEC"]/tbody/tr/td[2]')[0].text
    print(bairro)

    #pega os dados da cidade pelo X.PATH
    cidade = navegador.find_elements(By.XPATH, '//*[@id="resultado-DNEC"]/tbody/tr/td[3]')[0].text
    print(cidade)

    #pega os dados do cep pelo X.PATH
    cep = navegador.find_elements(By.XPATH, '//*[@id="resultado-DNEC"]/tbody/tr/td[4]')[0].text
    print(cep)

    #seleciona a sheet de dados
    sheetDadosParaImprimirEndereço = planilhaDadosEndereco['Dados']
    linhaCorrente = len(sheetDadosParaImprimirEndereço['A']) + 1
    
    colunaA = 'A' + str(linhaCorrente) #criando a variável para juntar A + a última linha ex: A2 + A3 + A4 +A5...
    colunaB = 'B' + str(linhaCorrente) #criando a variável para juntar B + a última linha ex: B2 + B3 + B4 +B5...
    colunaC = 'C' + str(linhaCorrente) #criando a variável para juntar C + a última linha ex: C2
    colunaD = 'D' + str(linhaCorrente) #criando a variável para juntar D + a última linha ex: D2

    #imprimindo as informações do site na planilha
    sheetDadosParaImprimirEndereço[colunaA] = rua #A2 + A3 + A4 +A5...
    sheetDadosParaImprimirEndereço[colunaB] = bairro #B2 + B3 + B4 +B5...
    sheetDadosParaImprimirEndereço[colunaC] = cidade #C2
    sheetDadosParaImprimirEndereço[colunaD] = cep #D2

#salvando o arquivo do excel com as novas informações
planilhaDadosEndereco.save(filename = nomeArquivoCEP)

#abrindo e apresentando na tela os arquivos com os dados
os.startfile(nomeArquivoCEP)
```
